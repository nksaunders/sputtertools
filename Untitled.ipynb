```python
import matplotlib.pyplot as pl
import numpy as np
import skope
import george
from george import kernels
% matplotlib inline
```

```python
star = skope.Target()
fpix, flux, ferr = star.GenerateLightCurve(12., ncadences = 100, roll = 1.)
```

```python
fig = pl.figure(figsize=(15,5))

t = np.linspace(0,99,len(flux))
ferr = np.sum(_.reshape((star.ncadences), -1), axis=1)

pl.errorbar(t,flux,yerr=ferr,fmt='k.')
```

```python
fnorm = flux - np.mean(flux)
```

```python
kernel = np.var(fnorm) * kernels.ExpSquaredKernel(0.5)
gp = george.GP(kernel)
```

```python
gp.compute(t, ferr)

pred, pred_var = gp.predict(fnorm, t, return_var=True)
```

```python
fig = pl.figure(figsize=(15,5))
pl.plot(t,pred,'r')
pl.fill_between(t, pred - np.sqrt(pred_var), pred + np.sqrt(pred_var),
                color="k", alpha=0.2)

pl.errorbar(t,fnorm,yerr=ferr,fmt='k.')
```
